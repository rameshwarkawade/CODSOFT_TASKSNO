```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

```python
df = pd.read_csv("Sample-Superstore.csv")
df.head()
```

```python
# Remove unnecessary column
df = df.drop(columns=['csvbase_row_id'])

# Display first 5 rows
df.head()
```

```python
# Convert date columns into datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Check updated data types
df.info()
```

```python
# Sales by Category

category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
category_sales.plot(kind='bar')

plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')

plt.xticks(rotation=0)
plt.show()
```

```python
# Display summary statistics of numerical columns
df.describe()
```

```python
# Display data types of all columns
df.dtypes
```

```python
# Total Sales by Category

category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
category_sales.plot(kind='bar', color=['skyblue','orange','green'])

plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.xticks(rotation=0)

plt.show()
```

# Observation
* Technology has the highest total sales.
* Furniture is the second-highest sales category.
* Office Supplies also contribute significantly but have lower total sales than Technology.

```python
# Total Profit by Category

category_profit = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
category_profit.plot(kind='bar', color=['green', 'skyblue', 'orange'])

plt.title('Total Profit by Category')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.xticks(rotation=0)

plt.show()
```

# Observation

* Technology generates the highest total profit.

* Office Supplies also contribute a significant amount of profit.

* Furniture has the lowest total profit among the three categories.










```python
# Total Sales by Region

region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
region_sales.plot(kind='bar', color=['royalblue', 'orange', 'green', 'red'])

plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.xticks(rotation=0)

plt.show()
```

# Observation

*   The West region records the highest total sales.
*   The East region also contributes significantly to overall sales.
*  The South region has the lowest total sales among all regions.






```python
# Sales by Segment
segment_sales = df.groupby('Segment')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
segment_sales.plot(kind='bar', color=['steelblue','orange','green'])

plt.title('Total Sales by Segment')
plt.xlabel('Segment')
plt.ylabel('Sales')
plt.xticks(rotation=0)

plt.show()
```

# Observation

* Consumer segment has the highest total sales.
* Corporate segment is the second-highest contributor.
* Home Office has the lowest total sales among the three segments.

```python
# Top 10 Products by Sales
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
top_products.plot(kind='bar', color='purple')

plt.title('Top 10 Products by Sales')
plt.xlabel('Product Name')
plt.ylabel('Total Sales')
plt.xticks(rotation=75)

plt.show()
```

# Observation

* The top 10 products contribute a significant share of total sales.
* A few products generate much higher sales than the others.
* These products can be prioritized for inventory planning and marketing strategies.

```python
# Total Sales by Region
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
region_sales.plot(kind='bar', color=['red','skyblue','green','orange'])

plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.xticks(rotation=0)

plt.show()
```

# Observation

* The West region records the highest total sales.
* The East region is the second-highest sales contributor.
* The South region has the lowest total sales among all regions.

```python
# Total Profit by Region

region_profit = df.groupby('Region')['Profit'].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
region_profit.plot(kind='bar', color='teal')

plt.title('Total Profit by Region')
plt.xlabel('Region')
plt.ylabel('Profit')
plt.xticks(rotation=0)

plt.show()
```

# Observation

*  The West region generates the highest total profit.
*  The East region also performs well in terms of profit.
*  The Central and South regions generate comparatively lower profits.

```python
# Correlation Heatmap

plt.figure(figsize=(6,4))

sns.heatmap(df[['Sales','Quantity','Discount','Profit']].corr(),
            annot=True,
            cmap='coolwarm',
            fmt='.2f')

plt.title('Correlation Heatmap')
plt.show()
```

# Observation
*  Sales and Profit have a positive correlation.
*  Discount has a negative relationship with Profit.
*  Quantity has a weak correlation with Sales and Profit.

# Conclusion

This Exploratory Data Analysis (EDA) project analyzed the Sample Superstore dataset to identify important business insights.

### Key Findings
- Technology generated the highest sales and profit.
- The West region recorded the highest overall sales and profit.
- The Consumer segment contributed the most to total sales.
- A few products generated a significant share of total revenue.
- Higher discounts generally reduced overall profit.
- Sales and Profit showed a positive correlation, while Discount had a negative correlation with Profit.

### Business Recommendations
- Focus on high-performing products and regions.
- Optimize discount strategies to improve profitability.
- Maintain sufficient inventory for top-selling products.
- Improve marketing efforts in lower-performing regions.
