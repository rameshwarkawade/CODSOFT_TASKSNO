## 📊 Bar Chart: Average Car Price by Brand

This chart shows the average selling price of cars for each brand.
It helps identify which brands have higher-priced vehicles.

```python
# Average Price by Brand

brand_price = df.groupby("Brand")["Price_USD"].mean().sort_values(ascending=False)

plt.figure(figsize=(10,6))
brand_price.plot(kind='bar', color='blue', edgecolor='black')

plt.title("Average Car Price by Brand")
plt.xlabel("Brand")
plt.ylabel("Average Price (USD)")
plt.xticks(rotation=45)
plt.show()
```

Insight:
• Toyota, Honda, and Tata have relatively higher average car prices in this dataset.

• This chart helps compare brand-wise pricing trends.

## 📈 Line Chart: Average Car Price by Manufacturing Year

This chart illustrates how the average selling price changes across manufacturing years.

```python
year_price = df.groupby("Year")["Price_USD"].mean()

plt.figure(figsize=(10,5))

plt.plot(
    year_price.index,
    year_price.values,
    marker='o',
    linewidth=2,
    color='green'
)

plt.title("Average Car Price by Manufacturing Year")
plt.xlabel("Year")
plt.ylabel("Average Price (USD)")
plt.grid(True)

plt.show()
```

Insight:
• The average selling price varies across manufacturing years.

• Newer models generally tend to have higher prices.:

## 🥧 Pie Chart: Car Brand Distribution

This chart displays the percentage distribution of different car brands in the dataset.

```python
# Car Brand Distribution

brand_count = df["Brand"].value_counts()

plt.figure(figsize=(8,8))

plt.pie(
    brand_count,
    labels=brand_count.index,
    autopct='%1.1f%%',
    startangle=90
)

plt.title("Car Brand Distribution")

plt.show()
```

Insight:
• This chart shows the proportion of each car brand in the dataset.

• It helps identify which brands appear most frequently.

## 📊 Histogram: Distribution of Car Prices

This histogram illustrates how car prices are distributed across all vehicles.

```python
plt.figure(figsize=(10,6))

plt.hist(
    df["Price_USD"],
    bins=15,
    color="skyblue",
    edgecolor="black"
)

plt.title("Distribution of Car Prices")
plt.xlabel("Price (USD)")
plt.ylabel("Number of Cars")

plt.show()
```

Insight:
• Most cars are concentrated within a particular price range.

• Extremely high-priced cars occur less frequently.

## 🔵 Scatter Plot: Mileage vs Car Price

This scatter plot analyzes the relationship between mileage and selling price.

```python
plt.figure(figsize=(10,6))

plt.scatter(
    df["Mileage_km"],
    df["Price_USD"],
    color="red",
    alpha=0.6
)

plt.title("Mileage vs Car Price")
plt.xlabel("Mileage (km)")
plt.ylabel("Price (USD)")

plt.show()
```

Insight:
• Cars with higher mileage generally tend to have lower selling prices.

• This chart helps understand the relationship between usage and market value.
